```python
import datetime

df['Date'] = pd.to_datetime(df['Date'], errors='coerce') # errors='coerce' конвертує помилкові значення в NaT

today = datetime.date.today()

def months_since(date):
    if pd.isna(date):
        return None
    return (today.year - date.year) * 12 + today.month - date.month

df['Months_Since'] = df['Date'].apply(months_since)
df.head()
```

```python
df.groupby('Store').agg(Average_sales = ("Sales", "mean"), Maximum_customers = ("Customers", "max") )
```

```python
df['Rank'] = df['Sales'].rank(method='dense', ascending=False)
df[df['Rank'] <= 3].sort_values('Rank')
```
